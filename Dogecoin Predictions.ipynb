```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import regression
sns.set()
plt.style.use('seaborn-whitegrid')
```

```python
doge_df = pd.read_csv("./Resources/coin_Dogecoin.csv")
```

```python
doge2020_df = doge_df[(doge_df["Date"] >= '2020-10-01')]
doge2020_df.head()
```

```python
doge2020_df.drop(columns = ['SNo'])
doge2020_df.set_index('Date')
```

```python
doge2020_df.dropna()
plt.figure(figsize=(10, 4))
plt.title("DogeCoin Price INR")
plt.xlabel("Date By Serial Number")
plt.ylabel("Close")
plt.plot(doge2020_df["Close"])
plt.show()
```

```python
doge2020_df
```

```python
from autots import AutoTS

# Set forecast length for amount of time you want to predict future value.
# Compare predicted Dogecoin for real data elsewhere.
model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple')
model = model.fit(doge2020_df, date_col='Date', value_col='Close', id_col=None)

prediction = model.predict()
forecast = prediction.forecast
print("DogeCoin Price Prediction")
print(forecast)
```
