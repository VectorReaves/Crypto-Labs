```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import regression
sns.set()
plt.style.use('seaborn-whitegrid')
```

```python
cosmos_df = pd.read_csv("./Resources/coin_Cosmos.csv")
```

```python
cosmos2020_df = cosmos_df[(cosmos_df["Date"] >= '2020-10-01')]
cosmos2020_df.head()
```

```python
cosmos2020_df.drop(columns = ['SNo'])
cosmos2020_df.set_index('Date')
```

```python
cosmos2020_df.dropna()
plt.figure(figsize=(10, 4))
plt.title("ATOM Price INR")
plt.xlabel("Date By Serial Number")
plt.ylabel("Close")
plt.plot(cosmos2020_df["Close"])
plt.show()
```

```python
from autots import AutoTS

# Set forecast length for amount of time you want to predict future value.
# Compare predicted Dogecoin for real data elsewhere.
model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple')
model = model.fit(cosmos2020_df, date_col='Date', value_col='Close', id_col=None)
 
prediction = model.predict()
forecast = prediction.forecast
print("ATOM Price Prediction")
print(forecast)
```
