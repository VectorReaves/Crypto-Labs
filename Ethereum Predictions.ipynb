```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import regression
sns.set()
plt.style.use('seaborn-whitegrid')
```

```python
ethereum_df = pd.read_csv("./Resources/coin_Ethereum.csv")
```

```python
ethereum2020_df = ethereum_df[(ethereum_df["Date"] >= '2020-10-01')]
ethereum2020_df.head()
```

```python
ethereum2020_df.dropna()
plt.figure(figsize=(10, 4))
plt.title("Ethereum Price INR")
plt.xlabel("Date By Serial Number")
plt.ylabel("Close")
plt.plot(ethereum2020_df["Close"])
plt.show()
```

```python
from autots import AutoTS

# Set forecast length for amount of time you want to predict future value.
# Compare predicted Dogecoin for real data elsewhere.
model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple')
model = model.fit(ethereum2020_df, date_col='Date', value_col='Close', id_col=None)
 
prediction = model.predict()
forecast = prediction.forecast
print("Ethereum Price Prediction")
print(forecast)
```
